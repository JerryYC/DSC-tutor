```python
import pandas as pd
import numpy as np
```

```python
canvas = pd.read_csv('canvas.csv').fillna(0)
```

```python
canvas
```

```python
canvas = canvas.loc[:,canvas.apply(lambda x: x.iloc[0] != '(read only)')]
```

```python
pa = [i for i in canvas.columns if 'PA' in i and 'quiz' not in i.lower()]
print(pa)
pa = canvas[pa].apply(lambda x: x/x.iloc[0]*100)
pa = pa.apply(lambda x: x.mean(),axis=1)
canvas['PA Total with Cap'] = pa.apply(lambda x: 100 if x> 100 else x)
canvas['PA Total'] = pa
```

```python
canvas.columns
```

```python
def drop_n_lowest(arr, num):
    arr = list(arr.values)
    deletions = 0
    while deletions < num:
        deletions +=1
        lowest_val = min(arr)
        arr.remove(lowest_val)
    return arr
rq = [i for i in canvas.columns if 'reading quiz' in i.lower() or 'zybook' in i.lower()]
print(rq)
rq = canvas[rq].apply(lambda x: x/x.iloc[0]*100)
rq = rq.apply(lambda x: np.mean(drop_n_lowest(x,3)),axis=1)
canvas['Reading Quiz Total'] = rq
```

```python
mid_s = [i for i in canvas.columns if 'midterm' in i.lower()]
print(mid_s)
mid = canvas[mid_s].apply(lambda x: x/x.iloc[0]*100)
mid = mid.apply(lambda x: x.mean(),axis=1)
canvas['Midterm Total'] = mid
```

```python
part = [i for i in canvas.columns if ('participation' in i.lower() and 'quiz' in i.lower() and 'reading' not in i.lower()) or 'quiz0' in i.lower()]
print(part)
pp = canvas[part].apply(lambda x: x/x.iloc[0]*100)
pp = pp.apply(lambda x: np.mean(drop_n_lowest(x,3)),axis=1)
canvas['Participation Total'] = pp
```

```python
di = [i for i in canvas.columns if 'discussion worksheet' in i.lower()]
print(di)
di = canvas[di].apply(lambda x: x/x.iloc[0]*100)
di = di.apply(lambda x: (x.sum()-x.min())/9,axis=1)
canvas['Discussion Total'] = di
```
