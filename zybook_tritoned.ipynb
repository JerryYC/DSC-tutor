```python
import pandas as pd
```

```python
name = "zybook_05_28"
```

```python
def run_code(name):
    zybook = pd.read_csv(name+'.csv')
    zybook_col = [i for i in zybook.columns if ('Participation' in i or 'Challenge' in i) and ('total' not in i)]
    print(zybook_col)
    zybook[name] = zybook[zybook_col].mean(axis=1)
    columns = [i for i in zybook.columns if 'zybook' in i]
    zybook = zybook[['Last name','First name','Primary email',name]]
    for column in columns:
        zybook[column] = zybook[column].apply(lambda x: 5 if x>80 else (3 if x>60 else (1 if x>40 else 0)))
    tritoned = pd.read_csv('canvas.csv')[['SIS Login ID','Student','SIS User ID']]
    zybook['id'] = zybook['Primary email'].apply(lambda x: x.split('@')[0])
    output = tritoned.merge(zybook,left_on='SIS Login ID', right_on = 'id',how='left')
    output[name] = output[name].fillna(0)
    output.to_csv('output.csv')
    output = tritoned.merge(zybook,left_on='SIS Login ID', right_on = 'id',how='outer')
    return output
```

```python
output = run_code(name)
```
